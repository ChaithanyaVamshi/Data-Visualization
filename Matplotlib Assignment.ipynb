```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline

url = 'https://raw.githubusercontent.com/Geoyi/Cleaning-Titanic-Data/master/titanic_original.csv'
titanic = pd.read_csv(url)
titanic.head()
```

## 1. Create a pie chart presenting the male/female proportion

```python
titanic['sex'].value_counts()  #Counts number of male and female
```

```python
proportion = titanic['sex'].value_counts() #created variable to store the count of male and female
labels = ['male','female']
plt.pie(proportion,labels=labels,shadow = True, autopct = '%1.1f%%',wedgeprops= {'edgecolor':'black'})
plt.show()
```

## 2. Create a scatterplot with the Fare paid and the Age, differ the plot color by gender

```python
titanic['sex'].isnull().sum() # Found 1 null value in sex column
```

```python
titanic['sex'].fillna('male',inplace = True)  # Hence filled the Nan value with male value since it has highest proportion
```

```python
colors = {'male':'red', 'female':'blue'}
plt.scatter(titanic['age'], titanic['fare'], alpha=0.75, s=80,c=titanic['sex'].apply(lambda x: colors[x]))
plt.xlabel('age')
plt.ylabel('fare')
plt.title('Fare paid and the Age')
plt.show()
```
